```python
import json
import pandas as pd
```

```python
df = pd.read_csv("../data/data-gov-india/industry-state-and-year-wise-startups-recognized-dpiit.csv")
```

```python
df.describe()
```

```python
yearly_df = df.groupby('Year')[['Count']].sum().reset_index()
display(yearly_df.head())
```

```python
state_df = df.groupby(by=['Year', 'State'])[['Count']].sum().reset_index()
display(state_df.head())
```

```python
yearly_df.to_dict(orient='records')
```

```python
state_df.to_dict(orient='records')
```

```python
with open('../data/data-gov-india/industry-state-and-year-wise-startups-recognized-dpiit.json', 'w') as f:
    json.dump(yearly_df.to_dict(orient='records'), f, indent=2)
```
