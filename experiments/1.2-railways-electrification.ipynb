```python
import json
import pandas as pd
```

```python
df = pd.read_csv("../data/data-gov-india/railways-electrified-route-km-1965-2014.csv")
```

```python
df.head()
```

```python
df2 = pd.read_csv("../data/data-gov-india/railways-electrified-route-km-2015-2026.csv")
```

```python
df2.head()
```

```python
df.rename(columns={'Total': 'RKM Electrified Cumulative'}, inplace=True)
df2.rename(columns={'RKM Cumulative': 'RKM Electrified Cumulative'}, inplace=True)
```

```python
combined_df = pd.concat([df[['FY', 'RKM Electrified Cumulative']], df2[['FY', 'RKM Electrified Cumulative']]])
```

```python
display(combined_df.head())
display(combined_df.tail())
```

```python
network_df = pd.read_csv("../data/data-gov-india/railways-total-network-km-1970-2013.csv")
```

```python
network_df.head()
```

```python
final_df = pd.merge(
    left=combined_df, right=network_df[['FY', 'Total Route Kms', 'Total Track Kms']], 
    on="FY"
    )
```

```python
final_df.head(2)
```

```python
final_df['Electrification %'] = final_df['RKM Electrified Cumulative']/final_df['Total Route Kms']
final_df.head(2)
```

```python
final_df.tail()
```

```python
with open('../data/data-gov-india/railways-network-stats-1971-2024.json', 'w') as f:
    json.dump(final_df.to_dict(orient='records'), f, indent=2)
```
